```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", context="talk")

# Load cleaned dataset
df = pd.read_csv("mpa_dr7_clean.csv")

print("Dataset shape:", df.shape)
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["logM_true"], bins=50, kde=True, color="steelblue")
plt.xlabel("log(Stellar Mass / Mâ˜‰)")
plt.ylabel("Number of Galaxies")
plt.title("Distribution of Stellar Mass")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["logSFR_true"], bins=60, kde=True, color="coral")
plt.xlabel("log(Star Formation Rate)")
plt.ylabel("Number of Galaxies")
plt.title("Distribution of Star Formation Rate")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x=df["logM_true"],
    y=df["logSFR_true"],
    s=8,
    alpha=0.3
)
plt.xlabel("log(Stellar Mass)")
plt.ylabel("log(SFR)")
plt.title("Galaxy Main Sequence and Quenched Population")
plt.tight_layout()
plt.show()
```

```python
df["u_r"] = df["u"] - df["r"]

plt.figure(figsize=(8,6))
sns.scatterplot(
    x=df["logM_true"],
    y=df["u_r"],
    s=8,
    alpha=0.3
)
plt.xlabel("log(Stellar Mass)")
plt.ylabel("u âˆ’ r Color")
plt.title("Colorâ€“Mass Diagram (Red vs Blue Galaxies)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["redshift"], bins=50, color="slategray")
plt.xlabel("Redshift")
plt.ylabel("Number of Galaxies")
plt.title("Redshift Distribution of Sample")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x=df["g"] - df["r"],
    y=df["u"] - df["g"],
    s=8,
    alpha=0.3
)
plt.xlabel("g âˆ’ r")
plt.ylabel("u âˆ’ g")
plt.title("Colorâ€“Color Diagram")
plt.tight_layout()
plt.show()
```
